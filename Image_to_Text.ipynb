```python
from transformers import pipeline
from PIL import Image
```

```python
caption_pipeline = pipeline("image-to-text", model="nlpconnect/vit-gpt2-image-captioning")
```

```python
import requests
from io import BytesIO
```

```python
image_url = "https://imgs.search.brave.com/WB6Yv1NZkahlDEEYgQfWm8mxI9bzFNW1Yd0rCpnx7o0/rs:fit:860:0:0:0/g:ce/aHR0cHM6Ly93d3cu/YWxqYXplZXJhLmNv/bS93cC1jb250ZW50/L3VwbG9hZHMvMjAy/NC8wOC9HZXR0eUlt/YWdlcy0yMTY2NDA3/MTcxLTE3MjQyNTkx/NTAuanBnP3Jlc2l6/ZT03NzAsNTEzJnF1/YWxpdHk9ODA"
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))
```

```python

import matplotlib.pyplot as plt
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
captions = caption_pipeline(image)
print(captions)
```
